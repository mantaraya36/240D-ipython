```python
signal = sin(linspace(0, 50, 512))
plot(signal)
```

```python
speaker_angle = 30 * (2 * pi/360)
distance = 1

scatter(distance*sin([speaker_angle, -speaker_angle]), distance*cos([speaker_angle, -speaker_angle]))
ylim(0, 1)
scatter(0,0, s=100, marker = 'o', c='r' )
grid()
```

```python
def linear_pan_gains(azimuth, speaker_angle):
    if azimuth > speaker_angle:
        azimuth = speaker_angle
    if azimuth < -speaker_angle:
        azimuth = -speaker_angle 
    gain_right = (azimuth + speaker_angle) / (2.0 * speaker_angle)
    gain_left = 1 - gain_right
    return gain_left, gain_right
```

```python
linear_pan_gains(0, 30)
```

```python
linear_pan_gains(-30, 30)
```

```python
linear_pan_gains(30, 30)
```

```python
linear_pan_gains(40, 30)
```

```python
gain_left, gain_right = linear_pan_gains(20, 30)
```

```python
sig_left = signal * gain_left
sig_right = signal * gain_right
```

```python
plot(sig_left)
plot(sig_right)
```
